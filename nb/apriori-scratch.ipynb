Apriori Algorithm Pseudo Code
Join Step: Ck is generated by joining Lk-1with itself

Prune Step:  Any (k-1)-itemset that is not frequent cannot be a subset of a frequent k-itemset
```
Pseudo-code:
Ck: Candidate itemset of size k

Lk: frequent itemset of size k

L1 = {frequent items};
for (k = 1; Lk !=0; k++) do begin
     Ck+1 = candidates generated from Lk;
    for each transaction t in database do
       increment the count of all candidates in Ck+1 that are contained in t
    Lk+1 = candidates in Ck+1 with min_support
    end
return = Lk;

```

````python
import logging
logging.basicConfig()
logger = logging.getLogger("foobar")
logger.setLevel(logging.DEBUG)

from itertools import combinations, permutations, product, accumulate

def read_csv_file(fname):
    """read a raw csv file as transactions"""
    file_iter = open(fname, 'r')
    raw_transactions = list()
    for line in file_iter:
        line = line.strip().rstrip(',')
        record = tuple(sorted(line.split(',')))
        raw_transactions.append(record)
        
    return raw_transactions

        
def generate_combinination_levels(tran_list, level):
    """generate keys that are used for subset checking"""
    """on each transaction"""
    results = list()
    for t in tran_list:
        #logger.debug(t)
        [ (logger.debug("new key: {0}".format(i)), results.append(i)) for i in combinations(t, level)]
        print("------")
        
    return sorted(set(results))


def gen_support_level(trans, items_keys):
    """for each key which can be a set find in transactions"""
    """how many contain the combination"""
    itemSet = dict()
    
    for key in items_keys:
        for t in trans:
            if set(key).issubset(t):
                logger.debug('is subset: {0}'.format(t))
                if (key) in itemSet:
                    itemSet[key] += 1
                else:
                    itemSet[key] = 1

    return itemSet;    
    
````
